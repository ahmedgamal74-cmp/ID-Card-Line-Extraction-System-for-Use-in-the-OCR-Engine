# Model Training

```python
from ultralytics import YOLO

model = YOLO("yolov8s-seg.pt")

results = model.train(
    data=data_yaml,
    imgsz=1024,
    epochs=100,
    batch=8,
    device=0,
    workers=2,
    cache=True
)
```

# Model Testing
